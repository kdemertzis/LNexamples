## Boosting algorithms 

Αποτελούνται από ένα σύνολο χαμηλής ακρίβειας ταξινομητών για τη δημιουργία ενός υψηλής ακρίβειας ταξινομητή. 
Ο boosting αλγόριθμος μπορεί να παρακολουθήσει το μοντέλο που απέτυχε την ακριβή πρόβλεψη καθώς και να αντιμετωπίσουν τα προβλήματα overffiting.
Οι boosting αλγόριθμοι έχουν αποκτήσει μαζική δημοτικότητα σε διαγωνισμούς επιστημονικών δεδομένων είναι οι AdaBoost (Adaptive Boosting) , Gradient Tree Boosting, XGBoost.  



### AdaBoost
Το Ada-boost ή το Adaptive Boosting είναι ένας από τους ensemble boosting classifier που προτάθηκε από τους Yoav Freund και Robert Schapire το 1996. Συνδυάζει πολλαπλούς ταξινομητές για να αυξήσει την ακρίβεια των ταξινομητών.

Το AdaBoost είναι μια επαναληπτική μέθοδος συνόλου. Ο ταξινομητής AdaBoost δημιουργεί έναν ισχυρό ταξινομητή, συνδυάζοντας πολλούς ταξινομητές με χαμηλή απόδοση, έτσι ώστε να δημιουργηθεί ταξινομητής υψηλής ακρίβειας.

Η βασική ιδέα πίσω από το Adaboost είναι η ρύθμιση των βαρών των ταξινομητών και η εκπαίδευση των training δεδομένων σε κάθε επανάληψη, έτσι ώστε να διασφαλίζεται ότι θα πραγματοποιηθούν ακριβείς προβλέψεις των ασυνήθιστων παρατηρήσεων. 

Οποιοσδήποτε αλγόριθμος μηχανικής μάθησης  μπορεί να χρησιμοποιηθεί ως ταξινομητής βάσης εάν δέχεται βάρη στο σετ εκπαίδευσης.

Το AdaBoost  λειτουργεί ιδιαίτερα καλά με τα δέντρα αποφάσεων.

![alt text](https://cdn-images-1.medium.com/max/800/1*bMvOaN1uNcaI9dvwH7McKw.png)

Το Adaboost θα πρέπει να πληροί δύο προϋποθέσεις:
- Ο ταξινομητής πρέπει να εκπαιδευτεί σε διάφορα διαβαθμισμένα παραδείγματα εκπαίδευσης. 
- Σε κάθε επανάληψη, προσπαθεί να παρέχει ένα καλό fit σε αυτά τα παραδείγματα, ελαχιστοποιώντας το σφάλμα που παρουσιάστηκε κατά την εκπαίδευση.


### Πώς λειτουργεί ο αλγόριθμος AdaBoost; 

-  Βήμα 1: Ο [Adaboost](https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.AdaBoostClassifier.html) επιλέγει τυχαία ένα υποσύνολο εκπαίδευσης και εκπαιδεύει ένα ταξινομητή βάσης.

-  Βήμα 2: Υπολογίζει το σταθμισμένο ποσοστό σφάλματος (e) του δέντρου αποφάσης.  Το σταθμισμένο ποσοστό σφάλματος (e) δείχνει το πλήθος των λανθασμένων συνολικά προβλέψεων στο σύνολο των ταξινομητών βάσης και χειρίζεται τις  λανθασμένες προβλέψεις διαφορετικά, σύμφωνα με το βάρος των δεδομένων του. Όσο μεγαλύτερο είναι το βάρος, τόσο περισσότερο σταθμίζεται το αντίστοιχο σφάλμα κατά τον υπολογισμό του e. 
                  
- Bήμα 3:  Υπολογίζει το βάρος του δέντρου απόφασης στο σύνολο των ταξινομητών βάσης.
                Βάρος του δέντρου= learning rate * log( (1 — e) / e)
                
Όσο υψηλότερο είναι το σταθμισμένο ποσοστό σφάλματος ενός δέντρου, τόσο λιγότερο θα συμμετέχει κατά τη διάρκεια της ψηφοφορίας ενώ αντίθετα όσο χαμηλότερο είναι το σταθμισμένο ποσοστό σφάλματος ενός δέντρου, τόσο περισσότερο θα συμμετέχει στη μεταγενέστερης ψηφοφορίας.
 
- Βήμα 4: Ενημερώνει τα βάρη των εσφαλμένα ταξινομημένων δεδομένων. Το βάρος κάθε δεδομένου θα αλλάξει μόνο αν το μοντέλο πήρε αυτό το δεδομένο λανθασμένο (το νέο βάρος του δεδομένου= παλαιό βάρος * np.exp (βάρος αυτού του δέντρου)). 
Όσο μεγαλύτερο είναι το βάρος του δέντρου, τόσο μεγαλύτερη βαρύτητα θα έχει το λανθασμένο δεδομένο από αυτό το δέντρο. Τα βάρη των δεδομένων κανονικοποιούνται αφού ενημερωθούν όλα τα εσφαλμένα ταξινομημένα δεδομένα. 

- Βήμα 5: Επαναλαμβάνει το Βήμα 1 (έως ότου επιτευχθεί ο αριθμός των δένδρων που θέτουμε για εκπαίδευση)

![alt text](https://cdn-images-1.medium.com/max/1000/1*MwIT3Gu-dhICzEou7he3OQ.png)

- Βήμα 6: Κάνει την τελική πρόβλεψη Το AdaBoost κάνει μια νέα πρόβλεψη προσθέτοντας το βάρος (κάθε δένδρου) πολλαπλασιάζοντας την πρόβλεψη (κάθε δένδρου). Προφανώς, το δέντρο με μεγαλύτερο βάρος θα έχει μεγαλύτερη δύναμη επιρροής στην τελική απόφαση.

Στη συνέχεια, θα πειραματιστούμε με τις μεθόδου συνόλου Adaboost & Random Forest



```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
# Load Library
from sklearn.datasets import make_blobs
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
```

```
X, y = make_blobs(n_samples=600, centers=2,
                  random_state=0, cluster_std=1.0)
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='rainbow');
```

```
#  Split the training test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```
#  Fit a Decision Tree model as comparison
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
y_pred_dt = clf.predict(X_test)
accDT=accuracy_score(y_test, y_pred_dt)
print(accDT)
```

```
# Fit a AdaBoost model with DT classifier, " compared to "Decision Tree model, accuracy go up by 10%
# Create and fit an AdaBoosted decision tree
bdt = AdaBoostClassifier(DecisionTreeClassifier(max_depth=1),
                         algorithm="SAMME",
                         n_estimators=500)

bdt.fit(X_train, y_train)
y_pred_ada = bdt.predict(X_test)
accAda=accuracy_score(y_test, y_pred_ada)
print(accAda)
```

```
# Import Support Vector Classifier
from sklearn.svm import SVC
svc=SVC(probability=True, kernel='linear')

bdt =AdaBoostClassifier(n_estimators=50, base_estimator=svc)
bdt.fit(X_train, y_train)
y_pred_svc = bdt.predict(X_test)
accAdaSVC=accuracy_score(y_test, y_pred_svc)
print(accAdaSVC)
```

```
#  Fit a Random Forest model, " compared to "Decision Tree model, accuracy go up by 5%
rlf = RandomForestClassifier(n_estimators=500, max_features="auto",random_state=0)
rlf.fit(X_train, y_train)
y_pred = rlf.predict(X_test)
accRF=accuracy_score(y_test, y_pred)
print(accRF)
```

```
print('Accuracy score\n')
print("Decision Tree", accDT, sep=" \t ")
print("\n")
print("Adaboost - DT", accAda, sep=" \t ")
print("\n")
print("Adaboost - SVC", accAdaSVC, sep=" \t ")
print("\n")
print("Random Forest", accRF, sep=" \t ")


### Plotting

# Decison tree

plot_colors = "br"
plot_step = 0.02
class_names = "AB"

plt.figure(figsize=(10, 5))

# Plot the decision boundaries
x_min, x_max = X_test[:, 0].min() - 1, X_test[:, 0].max() + 1
y_min, y_max = X_test[:, 1].min() - 1, X_test[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
plt.axis("tight")

# Plot the training points
for i, n, c in zip(range(2), class_names, plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1],
                c=c, cmap=plt.cm.Paired,
                s=20, edgecolor='k',
                label="Class %s" % n)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.legend(loc='upper right')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Decision Boundary Decision Tree')


## Adaboost DT

plot_colors = "br"
plot_step = 0.02
class_names = "AB"

plt.figure(figsize=(10, 5))

# Plot the decision boundaries
x_min, x_max = X_test[:, 0].min() - 1, X_test[:, 0].max() + 1
y_min, y_max = X_test[:, 1].min() - 1, X_test[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

Z = bdt.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
plt.axis("tight")

# Plot the training points
for i, n, c in zip(range(2), class_names, plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1],
                c=c, cmap=plt.cm.Paired,
                s=20, edgecolor='k',
                label="Class %s" % n)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.legend(loc='upper right')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Decision Boundary Adaboost DT')

plt.show()

## Adaboost SCV

plot_colors = "br"
plot_step = 0.02
class_names = "AB"

plt.figure(figsize=(10, 5))

# Plot the decision boundaries
x_min, x_max = X_test[:, 0].min() - 1, X_test[:, 0].max() + 1
y_min, y_max = X_test[:, 1].min() - 1, X_test[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

Z = bdt.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
plt.axis("tight")

# Plot the training points
for i, n, c in zip(range(2), class_names, plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1],
                c=c, cmap=plt.cm.Paired,
                s=20, edgecolor='k',
                label="Class %s" % n)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.legend(loc='upper right')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Decision Boundary Adaboost SCV')

plt.show()
## Random forest

plot_colors = "br"
plot_step = 0.02
class_names = "AB"

plt.figure(figsize=(10, 5))

# Plot the decision boundaries
x_min, x_max = X_test[:, 0].min() - 1, X_test[:, 0].max() + 1
y_min, y_max = X_test[:, 1].min() - 1, X_test[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, plot_step))

Z = rlf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
plt.axis("tight")

# Plot the training points
for i, n, c in zip(range(2), class_names, plot_colors):
    idx = np.where(y == i)
    plt.scatter(X[idx, 0], X[idx, 1],
                c=c, cmap=plt.cm.Paired,
                s=20, edgecolor='k',
                label="Class %s" % n)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.legend(loc='upper right')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Decision Boundary Random Forest')

plt.show()
```
