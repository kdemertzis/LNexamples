```python
import pandas as pd

df = pd.read_csv('/content/BATADAL.csv')
```

```python
from ctgan import CTGAN

# Instantiate a CTGAN model
ctgan = CTGAN()

# Train the model on the real dataset
ctgan.fit(df)

# Generate a synthetic dataset with the same structure and statistical properties as the real dataset
synthetic_data = ctgan.sample(len(df))
```

```python
synthetic_data.to_csv('my_synthetic_data.csv', index=False)
```
